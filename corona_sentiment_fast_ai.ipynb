```
import pandas as pd
import torch
from fastai.text.all import *
torch.cuda.set_device(0)
```

```
data_dir = "/content/gdrive/My Drive/nlp_exp/data/corona_tweets/"
```

```
train_data_raw = pd.read_csv(data_dir + "Corona_NLP_train.csv",  encoding = "ISO-8859-1")
test_data_raw = pd.read_csv(data_dir + "Corona_NLP_test.csv")
```

```
data_raw = pd.concat([train_data_raw, test_data_raw]).reset_index(drop = True)
label_encoding = {'Extremely Negative': 0,  'Negative': 1,  'Neutral': 2, 'Positive': 3, 'Extremely Positive': 4} 
data_raw = data_raw[["OriginalTweet", "Sentiment"]]
```

```
display(data_raw)
```

```
dls = TextDataLoaders.from_df(data_raw, valid_pct = 0.2, text_col = 0, label_col = 1, device = 0)
```

```
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
```

```
learn.fine_tune(8, 3e-2)
```

```
learn.show_results()
```
